```python
def plotthecurve(xdata,ydata,xlbl,ylbl):    
    f, ax = plt.subplots(1)
    f.set_size_inches(10,10) 
    plt.xlabel(xlbl)
    plt.ylabel(ylbl)
    ax.plot(xdata, ydata,'bo-')
    ax.set_ylim(ymin=0)
    ax.set_xlim(xmin=0) 
    
    major_ticks = np.arange(0, 101, 20)                                              
    minor_ticks = np.arange(0, 101, 5)                                               

    ax.set_xticks(np.arange(0,max(xdata)*1.1,max(xdata)/10)) 
    ax.set_xticks(np.arange(0,max(xdata)*1.1,max(xdata)/50),minor=True)   
    #ax.set_xticks(minor_ticks, minor=True)
    ax.set_yticks(np.arange(0,max(ydata)*1.1,max(ydata)/10))
    ax.set_yticks(np.arange(0,max(ydata)*1.1,max(ydata)/50),minor=True)
    #ax.set_yticks(major_ticks)                                                       
    #ax.set_yticks(minor_ticks, minor=True)  
    ax.grid(which='both')                                                            

    # Grid Bar Opacity:                               
    ax.grid(which='minor', alpha=0.3)                                                
    ax.grid(which='major', alpha=0.8) 

    plt.show(f)
```

```python
xdata = [x*0.001 for x in [1,4,7,10,13]]
ydata = [.17,.43,.72,1.02,1.35]
plotthecurve(xdata,ydata,'Voltage','Current (mA)')
```

## Table - 2: Current Voltage Characteristics
Intensity $I_2$ = 150V

```python
xdata = [x for x in [16,64,144,256,400]]
I = [0.01,0.03,0.15,0.18,0.64]
V = [0.010,0.021,0.093,0.110,0.391]
ydata = [i*v for i,v in zip(I,V)]
for a,v,i,p in zip(xdata,V,I,ydata):
    print("%.2f  %0.2f  %0.2f  %0.2f" %( a,v,i,p))
plotthecurve(xdata,ydata,'Area ($mm^2$)','Power (mW)')
```
